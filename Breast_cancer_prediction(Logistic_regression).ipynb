Importing the dependencies


```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# importing the datset from the skelarn library here no need for csv and all
```

Data Collection and Preprocessing

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading a data to panda dataset
data_frame=pd.DataFrame(breast_cancer_dataset.data, columns= breast_cancer_dataset.feature_names )
```

```python
data_frame.head()
```

```python
# adding the target column

data_frame['label']= breast_cancer_dataset.target
```

```python
data_frame.tail()
```

```python
#number of rows and columns in the dataset
data_frame.shape
```

```python
data_frame.info()
```

```python
#checking for missing values

data_frame.isnull().sum()
```

```python
# statistical meausres about the data
data_frame.describe()
```

```python
# checking the labels if yes /no
data_frame['label'].value_counts()
```

1->no tumor
0-> tumor

```python
data_frame.groupby('label').mean()
```

seperating the features and target

```python
# input are x and lable colun is y
X=data_frame.drop(columns='label',axis=1)
Y=data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Split the dataset into training and test data

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2,random_state=2 )
```

```python
print(X_train)
```

```python
print(Y_train)
```

```python
print(X_test)
```

```python
print(Y_test)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training (Logistic Regression)

```python
model= LogisticRegression()
```

```python
# training the model using training data
model.fit(X_train,Y_train)
```

Model Evaluation

```python
# accuracy
X_train_prediction = model.predict(X_train)
training_data_accuracy= accuracy_score(Y_train,X_train_prediction)
```

```python
print( training_data_accuracy )
```

```python
X_test_prediction= model.predict(X_test)
test_data_accuracy=accuracy_score(Y_test,X_test_prediction)
```

```python
print(test_data_accuracy )
```

Building a prediction a system


```python
input_data = (13.54,14.36,87.46,566.3,0.09779,0.08129,0.06664,0.04781,0.1885,0.05766,0.2699,0.7886,2.058,23.56,0.008462,0.0146,0.02387,0.01315,0.0198,0.0023,15.11,19.26,99.7,711.2,0.144,0.1773,0.239,0.1288,0.2977,0.07259)

# change the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for one datapoint
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0] == 0):
  print('The Breast cancer is Malignant/tumor')

else:
  print('The Breast Cancer is Benign/no tumor')

```
